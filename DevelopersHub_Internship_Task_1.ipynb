**Step 1: Import Required Libraries**

```python
# pandas is used to work with datasets
import pandas as pd

# seaborn and matplotlib are used for visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

**Step 2: Load the Dataset**

```python
# Load Iris dataset (already available in seaborn)
iris_df = sns.load_dataset("iris")
```

**Step 3: Understand the Dataset**

Dataset Shape

```python
print("Dataset Shape:", iris_df.shape)
```

Columns in Dataset

```python
print("Column Names:", iris_df.columns)
```

First 5 Rows

```python
iris_df.head()
```

**Step 4: Dataset Information**

```python
# Shows data types and checks for missing values
iris_df.info()
```

**Step 5: Statistical Summary**

```python
# Shows mean, min, max, std, etc.
iris_df.describe()
```

**Step 6: Data Visualization**

Scatter Plot (Relationship Between Features)

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(
    data=iris_df,
    x="sepal_length",
    y="sepal_width",
    hue="species"
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

Histogram (Distribution of Values)

```python
iris_df.hist(figsize=(9, 7))
plt.suptitle("Histogram of Iris Features")
plt.show()
```

Box Plot (Outlier Detection)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris_df.drop("species", axis=1))
plt.title("Box Plot of Features")
plt.show()
```
